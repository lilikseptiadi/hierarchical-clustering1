```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn.datasets import make_blobs
```

```python
dataX, dataY = make_blobs(random_state=0, n_samples=1000)
print("dataX", dataX)
print("dataY", dataY)
```

```python
import matplotlib.pyplot as gambar
```

```python
gambar.plot(dataX[:,0],dataX[:,1],'ro')
gambar.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=3,linkage="average",affinity='manhattan')
```

```python
clustering.fit(dataX)
```

```python
gambar.scatter(dataX[:,0],dataX[:,1],c=clustering.labels_, marker='o')
gambar.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=3,linkage="complete",affinity='manhattan')
```

```python
clustering.fit(dataX)
```

```python
gambar.scatter(dataX[:,0],dataX[:,1],c=clustering.labels_, marker='o')
gambar.show()
```
